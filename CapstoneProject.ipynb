```python
#Dropping NaN Values
df.dropna(subset=['Y_COORD_CD','X_COORD_CD','Latitude','Longitude','ARREST_BORO','OFNS_DESC'], inplace=True)
```

```python
#Using replace function

df1= df.replace({'HARRASSMENT 2': 'HARASSMENT', 
                'ESCAPE 3': 'ESCAPE',
                'ASSAULT 3 & RELATED OFFENSES': 'ASSAULT & RELATED OFFENSES',
                'CRIMINAL MISCHIEF & RELATED OF': 'CRIMINAL MISCHIEF',
                'OFF. AGNST PUB ORD SENSBLTY &': 'OFFENSES AGAINST PUBLIC ORDER/ADMINISTRATION',
                'F': 'FEMALE','M': 'MALE','K':'BROOKLYN','Q': 'QUEENS','B':'BRONX','S':'STATEN ISLAND'}).copy()
```

```python
#Checking with the new replacement
df1.head(15)
```

```python
#The new variable names and data types
df1.dtypes
```

```python
#Analyzing rape crimes
df2=df1[df.OFNS_DESC.str.contains('FELONY SEX CRIMES|RAPE')]
```

```python
#Vizualization of Rape Crimes by Gender
df2['PERP_SEX'].value_counts().iloc[:3].plot(kind="bar", title = 'Rape Crimes by Gender', rot=0)
```

```python
#Vizualization of Rape Perps by Race
df2['PERP_RACE'].value_counts().iloc[:7].plot(kind="barh", title = 'Rape Crime Perps by Race')
```
